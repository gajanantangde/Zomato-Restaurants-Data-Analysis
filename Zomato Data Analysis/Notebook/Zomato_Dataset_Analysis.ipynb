```python
ratings = (
    df
    .groupby(['Aggregate rating', 'Rating color', 'Rating text'])
    .size()
    .reset_index(name='Rating Counts')
)
```

```python
ratings
```

## Conclusions 
1. When Rating is between 4.5 to 4.9 ---> Excellent
2. When ratings are between 4.0 to 4.5 ---> very good
3. When Rating is between 3.5 to 3.9 ----> Good
4. When Rating is between 2.5 to 3.5 ----> Average
5. When Rating is between 1.5 to 2.5 ----> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Counts',hue='Rating color', data=ratings,palette=['gray','red','orange','yellow','green','skyblue'])
plt.savefig(
    r"C:\Users\Lenovo\Documents\Data Science And AI 21 April Batch\Batch 01-08-2025\EDA\Zomato Data Analysis\Charts' Images\aggregate_rating_bar.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```
